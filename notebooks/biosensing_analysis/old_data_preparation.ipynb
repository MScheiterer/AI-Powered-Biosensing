Choose an image batch by setting target_image and it will be collected from all folders in folder_names and added to the data dictionary compatible with the analysis framework

```python
import os
import shutil
import json

# Modify this with the actual path to your data
base_path = "..\data\Raw\Water-Nacl"

# Folders names - adjust if necessary
folder_names = [
    "Chip2-NaCl 0%",
    "Chip2-NaCl 1.3%",
    "Chip2-NaCl 2.6%",
    "Chip2-NaCl 3.9%",
    "Chip2-NaCl 5.2%",
    "Chip2-NaCl 6.5%",
    "Chip2-NaCl 10.4%",
    "Chip2-NaCl 14.7%",
    "Chip2-NaCl 19%"
]

# Set the specific image name you're looking for
target_image = "300-0,75-1x-1.jpg"

# Collect paths to that image from each folder
data = {target_image: []}

for folder in folder_names:
    full_image_path = os.path.join(base_path, folder, target_image)
    if os.path.exists(full_image_path):
        data[target_image].append({
                    "file_path": full_image_path,
                    "nacl_percentage": folder.split(' ')[-1].strip('%')
                    })
    else:
        print(f"⚠️ Image not found in folder: {folder}")

print(json.dumps(data, indent=4))


destination_folder = "..\data\Raw\Water-Nacl\eval_images"

# os.makedirs(destination_folder, exist_ok=True)

# for file_name, image_batch in data.items():
#     for image in image_batch:
#         batch_name = os.path.basename(image["file_path"]).split("-")
#         new_file_name = f"{batch_name[0]}-{batch_name[1]}-{batch_name[2]}-{image['nacl_percentage']}%-{batch_name[-1]}"
    
#         src_file = image["file_path"]
#         dest_file = os.path.join(destination_folder, new_file_name)

#         if os.path.exists(src_file):
#             shutil.copy(src_file, dest_file)
#             print(f"Copied: {file_name}")
#         else:
#             print(f"❌ File not found: {file_name}")
```
